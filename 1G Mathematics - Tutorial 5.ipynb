### Activity 1: Food webs & Networks

You will be connecting food web descriptions to networks and corresponding adjacency matrices. 
Recall that adjacency matrices of *undirected graphs* are symmetric. Adjacency matrices of directed networks are not necessarily symmetric, instead. Indeed, they
have nonzero entry $A_{ij} \  (\neq A_{ji})$ if the node $i$ is connected to the node $j$ with a directed edge that goes from $i$ to $j$. 

To help you get started, in the cell below we import necessary packages and defined two functions *food_web* and *plot_food_web*.

- *food_web* takes in the adjacency matrix A and the species present and prints out which species eat what.

- *plot_food_web* takes in the adjacency matrix A and the species present and plots the food web. 

**You don't need to know how this code works.**
But you are welcome to play around with it if you are interested.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def food_web(A, species):
    for i, predator in enumerate(species):
        prey_list = [
            f"{species[j]} ({A[j][i]*100}%)"
            for j in range(len(species)) if A[j][i] > 0
        ]
        if prey_list:
            print(f"{predator} eat {', '.join(prey_list)}")
        else:
            print(f"{predator} eat nothing")

def plot_food_web(A, species):
    symmetric = np.array_equal(A, A.T)
    if symmetric:
        G = nx.Graph()  # undirected if symmetric
    else:
        G = nx.DiGraph()
    
    # Add weighted edges
    for i, predator in enumerate(species):
        for j, prey in enumerate(species):
            if A[i, j] > 0:
                G.add_edge(predator, prey, weight=A[i, j])
                
    plt.figure(figsize=(8, 6)) 
    
    pos = nx.spring_layout(G, seed=42, k=2.2)
    pos = {k: (v[0]*3, v[1]*3) for k, v in pos.items()}
    
    # Draw graph
    nx.draw(G, pos,with_labels=True,node_size=2300,node_color="lightblue",font_size=10,arrows=True,arrowsize=25,width=1.5)

    # Draw edge weight labels 
    if not symmetric:
        edge_labels = nx.get_edge_attributes(G, "weight")
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, label_pos=0.65,rotate=False)

    plt.axis("off")
    
    plt.show()
```

We will now give you an **example** food web, where:
- Sea Otters are eaten by Sharks. 
- Sea Otters eat Sea Urchins.

We define a vector called species, containing all species in the food web, and the adjacency matrix that corresponds to this food web.

We then print and plot the food web using the above defined *food_web* and *plot_food_web* functions.

```python
species = ["Sea Otters", "Sharks", "Sea Urchins"]

# Adjacency matrix 
A = np.array(
    [[0, 1, 0],  
     [0, 0, 0],  
     [1, 0, 0]])   

food_web(A, species)

plot_food_web(A, species)
```

The corresponding undirected network and adjacency matrix look as follows:

```python
# Adjacency matrix 
A = np.array(
    [[0, 1, 1],  
     [1, 0, 0],  
     [1, 0, 0]])   

plot_food_web(A, species)
```
