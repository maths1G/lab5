### Extra activity: Dynamically Remove and Add Species to your Food Web

To help you get started, in the cell below we import necessary packages and defined four functions *food_web*, *plot_food_web*, *remove_species* and *update_relationship*.

- *food_web* takes in the adjacency matrix A and the species present and prints out which species eat what.

- *plot_food_web* takes in the adjacency matrix A and the species present and plots the food web. 

- *remove_species* takes in the adjacency matrix A and the species present and allows the user to insert the name of a species to be removed from the foodweb. The new food web is then plotted.

- *update_relationship* takes in the adjacency matrix A and the species present and allows the user to insert the name of a predator-prey relationship to be added to the food web. The new food web is then plotted.

**You don't need to know how this code works.**
But you are welcome to play around with it if you are interested.


```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def food_web(A, species):
    for i, predator in enumerate(species):
        prey_list = [
            f"{species[j]} ({A[j][i]*100}%)"
            for j in range(len(species)) if A[j][i] > 0
        ]
        if prey_list:
            print(f"{predator} eat {', '.join(prey_list)}")
        else:
            print(f"{predator} eat nothing")

def plot_food_web(A, species):
    symmetric = np.array_equal(A, A.T)
    if symmetric:
        G = nx.Graph()  # undirected if symmetric
    else:
        G = nx.DiGraph()
    
    # Add weighted edges
    for i, predator in enumerate(species):
        for j, prey in enumerate(species):
            if A[i, j] > 0:
                G.add_edge(predator, prey, weight=A[i, j])
                
    plt.figure(figsize=(8, 6)) 
    
    pos = nx.spring_layout(G, seed=42, k=2.2)
    pos = {k: (v[0]*3, v[1]*3) for k, v in pos.items()}
    
    # Draw graph
    nx.draw(G, pos,with_labels=True,node_size=2300,node_color="lightblue",font_size=10,arrows=True,arrowsize=25,width=1.5)

    # Draw edge weight labels 
    if not symmetric:
        edge_labels = nx.get_edge_attributes(G, "weight")
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, label_pos=0.65,rotate=False)

    plt.axis("off")
    
    plt.show()
    
def remove_species(Aint,species_int):
    n = input("Species to remove: ")
    species_int = species_int.copy()  
    Aint = Aint.copy()         

    if n in species_int:
        i = species_int.index(n)
        Aint = np.delete(Aint, i, 0) 
        Aint = np.delete(Aint, i, 1)  
        species_int.pop(i)
        print(f"Removed species: {n}")
        plot_food_web(Aint, species_int)
    else:
        print("Species not found!")
    
def update_relationship(Aint, species_int):
    species_int = species_int.copy()
    Aint = Aint.copy()
    
    predator = input("Enter predator species: ")
    prey = input("Enter prey species: ")

    # Add predator if new
    if predator not in species_int:
        species_int.append(predator)
        for row in Aint:
            row.append(0)
        Aint.append([0]*len(species_int))

    # Add prey if new
    if prey not in species_int:
        species_int.append(prey)
        for row in Aint:
            row.append(0)
        Aint.append([0]*len(species_int))

    i, j = species_int.index(predator), species_int.index(prey)

    # Ask user for weight of this prey in predator's diet
    while True:
        try:
            new_weight = float(input(f"What fraction (%) of {predator}'s diet is {prey}? (0-100): "))
            if 0 <= new_weight <= 100:
                new_weight /= 100  # convert to 0-1 scale
                break
            else:
                print("Enter a value between 0 and 100.")
        except ValueError:
            print("Invalid: Please enter a number between 0 and 100.")

    # Get current diet weights for this predator 
    current_weights = [Aint[k][i] for k in range(len(species_int))]

    # Sum of current diet excluding the prey being updated
    other_sum = sum(current_weights) - current_weights[j]

    # Rescale other prey proportionally if total exceeds 1
    if other_sum + new_weight > 1:
        scaling_factor = (1 - new_weight) / other_sum if other_sum > 0 else 0
        for k in range(len(species_int)):
            if k != j:
                Aint[k][i] *= scaling_factor

    # Set new weight for the prey
    Aint[j][i] = new_weight

    print(f"Updated diet for {predator}:")
    for k, sp in enumerate(species_int):
        if Aint[k][i] > 0:
            print(f" - {sp}: {Aint[k][i]*100:.1f}%")

    plot_food_web(Aint, species_int)
```
